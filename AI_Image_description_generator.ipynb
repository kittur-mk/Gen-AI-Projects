```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

# Load BLIP processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
def generate_caption(image):
    inputs = processor(images=image, return_tensors="pt")
    out = model.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption
```

```python
import gradio as gr

custom_css = """
body {
    background: linear-gradient(to right, #0f0c29, #302b63, #24243e);
    font-family: 'Courier New', monospace;
    color: #39ff14;
}

h1, p {
    text-align: center;
    font-family: 'Press Start 2P', monospace;
    font-size: 1.5em;
    color: #00ffff;
    text-shadow: 0 0 5px #ff00ff, 0 0 10px #ff00ff, 0 0 20px #ff00ff;
}

.gr-button {
    background-color: #8e2de2 !important;
    color: #ffffff !important;
    border-radius: 5px;
    font-weight: bold;
    font-family: 'Courier New', monospace;
    box-shadow: 0px 0px 8px #ff00ff;
    border: 2px solid #39ff14;
    transition: 0.4s ease;
}

.gr-button:hover {
    background-color: #4a00e0 !important;
    box-shadow: 0px 0px 16px #ff00ff;
}

.gr-image-preview {
    border-radius: 10px;
    box-shadow: 0 0 20px #00ffcc;
    border: 3px dashed #39ff14;
}

.gr-textbox textarea {
    background-color: #000000;
    color: #00ffcc;
    font-family: 'Courier New', monospace;
    border: 2px solid #39ff14;
    box-shadow: 0 0 10px #39ff14;
}

footer {
    display: none !important;
}
"""


with gr.Blocks(css=custom_css) as demo:
    gr.Markdown("<h1>🤗 MK's AI Image Caption Generator</h1>")
    gr.Markdown("✨ Upload an image and let the magic AI describe it!")

    with gr.Row():
        image_input = gr.Image(type="pil", label="Upload Image")
        caption_output = gr.Textbox(label="📝 AI Caption", lines=2)

    generate_btn = gr.Button("🎬 Generate Caption")


    gr.Examples(
        examples=[
            ["dog.jpg"],
            ["beach.jpg"],
            ["food.jpg"]
        ],
        inputs=[image_input]
    )

    generate_btn.click(fn=generate_caption, inputs=image_input, outputs=caption_output)

demo.launch()
```
